- hist

```python
plt.figure()
sb.histplot(data = origin, x = '월수입', bins = 5)
plt.grid()
plt.title("월 수입 히스토그램")
plt.show()
```

- 히스그램의 커널 밀도 표시

- 모집단 : 통계를 목표로 하는 전체 집단, 일반적으로 모집단의 정확한 통계는 전수조사가 아니면 현실적으로 낼 수 없음
  - 모수 : 모집단의 통계 지표들 (평균 등)
- 표본집단 : 모집단의 통계 지표를 확인하기 위해 대표로 선정한 집단
  - 비모수적 방법 : 관측한 값이 특정 확률분포를 따르지 않거나 모집단에 대한 정보가 없을 경우 실시하는 검정방법

```python
plt.figure()
sb.histplot(data = origin, x = '월수입', bins = 5, kde = True)
plt.grid()
plt.title("월 수입 히스토그램")
plt.show()
```

```python
df = pd.read_csv("./data/penguins.csv")
```

```python
df.head()
```

```python
sb.pairplot(df)
plt.show()
```

```python
df.info()
```

```python
sb.scatterplot(data = df, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species')
plt.grid()
plt.show()
```

```python
df2 = df.groupby(['species', 'island']).count()[['sex']].reset_index()
df2.rename(columns = {'sex' : 'count'}, inplace = True)
```

```python
df2
```

```python
sb.barplot(data = df2, x = 'island', y = 'count', hue = 'species')
plt.grid()
plt.legend(loc = 'upper right')
plt.show()
```

```python
sb.regplot(data = df, x = 'bill_length_mm', y = 'bill_depth_mm')
plt.grid()
plt.show()
```

```python
df.head()
```

```python
for i in df['species'].unique():
    sb.regplot(data = df[df['species'] == i], x = 'bill_length_mm', y = 'bill_depth_mm', label = i, ci = 0)
plt.grid()
plt.legend()
plt.show()
```

```python
df2 = pd.read_csv("./data/tips.csv")
```

```python
df2.head()
```

```python
df2.shape
```

```python
sb.regplot(data = df2, x = 'total_bill', y = 'tip')
plt.grid()
plt.show()
```

```python
df2_2 = df2.groupby('size').mean(numeric_only = True)
df2_2.reset_index(inplace = True)
```

```python
sb.barplot(data = df2_2, x = 'size', y = 'tip')
plt.grid()
plt.show()
```

```python
for i in df2['sex'].unique():
    sb.regplot(data = df2[df2['sex'] == i], x = 'total_bill', y = 'tip', label = i, ci = 0)
plt.legend()
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 6))
for i, v in enumerate(df2['sex'].unique()):
    sb.regplot(ax = ax[i], data = df2[df2['sex'] == v], x = 'total_bill', y = 'tip', label = v, ci = 0)
    ax[i].grid()
    ax[i].set_title(v)
    ax[i].set_ylim(df2['tip'].min() - 0.5, + df2['tip'].max() + 0.5)
    ax[i].set_xlim(df2['total_bill'].min() - 1, + df2['total_bill'].max() + 1)

plt.show()
```
