- 인공지능 -> 머신러닝 -> 딥러닝
  - 각 개념은 하위 카테고리로 분류

- 강인공지능 : 스스로 사고하고 학습하는 인공지능
- 약인공지능 : 머신러닝등 현대 대부분의 모든 인공지능이 해당, 스스로 학습하지는 못함

- 머신러닝 : 데이터를 이용하여 컴퓨터가 어떤 지식이나 패턴을 학습
- 딥러닝 : 머신러닝의 방법 중 하나로 인공신경망을 쌓아 학습

- 머신러닝은 입력한 데이터와 결과를 입력받아 규칙을 찾는것을 목적으로 함
  - 경사하강법 : 입력한 데이터/결과와 실제 데이터를 여러번 비교하여 그 오차를 점점 좁혀나가는 기법

- 머신러닝의 알고리즘 : 연구를 통해 이미 만들어진 공식을 프로그래밍화 시켜 생성된 함수
  - 지도학습 : 입력 데이터, 출력 데이터를 알려주고 그 사이의 연관성을 찾아내는 방식
    - 회귀분석 : 출력 데이터가 연속형
    - 분류분석 : 출력 데이터가 명목형(0과 1의 분류 : 이진분류)
  - 비지도학습 : 입력 데이터만 존재
    - 군집화 : 유사한 데이터끼리 묶음
      - 군집을 몇가지로 나눌 것인지는 사람이 정해줌
    - 패턴검색 : 데이터의 패턴을 분석
    - 차원축소 : 변수의 갯수를 줄이고 요약 (탐색적 데이터 분석으로 분류하기도 함)

- 앙상블 : 다수의 머신러닝 알고리즘을 조합하여 성능을 개선하는 방식

- 강화학습 : 환경에 대한 정보는 주지 않고 성공 / 실패를 알고리즘이 경험하고 스스로 결정을 내릴 수 있게 학습함

- 사이킷 런 : 지도학습, 비지도학습을 위한 다양한 모델 제공, 시각화, 검증 도구 등 다양한 기능이 있으나 딥러닝은 없음
  - 피처(feature) : 데이터에서의 열 개념

```python
from sklearn.utils import all_estimators
import pandas as pd
```

```python
filterType = ['classifier', 'regressor', 'cluster', 'transformer']
mylist = []

for f in filterType:
    allAlgorithms = all_estimators(type_filter=f)

    for name, package in allAlgorithms:
        cls = str(package)
        x = cls.find("'")
        y = cls.rfind("'")
        clsName = cls[x+1:y]

        if clsName.find("._") > -1:
            continue

        mylist.append({"type": f, "name": name, "class": clsName})

df = pd.DataFrame(mylist)
df.sort_values(['type', 'name'])
```
