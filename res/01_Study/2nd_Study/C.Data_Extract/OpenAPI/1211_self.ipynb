- HTTP 메서드 POST
  - 웹에 접속하여 데이터를 요청할 때 넘기는 파라미터 방식(querystring)은 url에 노출되므로 보안에 취약하고 1024바이트까지만 정보를 담을 수 있음
    - 파라미터를 쿼리스트링으로 넘기는 방식은 get 방식이라고 함 (session.get)
  - header에 실어서 넘기는 POST 방식은 보안에서는 상대적으로 안전하며 약 20MB정도까지 전송 가능, 대신 소스코드가 복잡해짐
    - 변수를 헤더에 넣는 방식은 post 방식이라고 함 (session.post)

- curl : 윈도우 명령어로, 명령프롬프트 기반으로 작동하는 url 페이지
  - -H로 표현되는 헤더를 포함시켜서 결과를 전달받을 수 있음 (파파고 : 클라이언트id, secret 정보)
  - -d는 전달하는 매개변수 (파파고 : 번역 대상 언어 및 텍스트)

- post 방식에서는 전달할 파라미터를 url이 아닌 data라는 별도의 매개변수에 딕셔러리 형태로 담아서 전달

- 리스트나 딕셔너리는 mutable이라서 함수 내에서 global 선언을 하지 않더라도 얕은 복사로 처리되어 함수 외부의 컬렉션이 영향을 받음

```python
def plus(x, y):
    return x + y
```

```python
def something(a, b, cb):
    print("%d와 %d의 연산 결과를 생성합니다..." % (a, b))
    z = cb(a, b)
    return z
```

```python
something(100, 200, plus) # 파라미터로 함수를 사용
```

- 단일 함수로 여러가지 명령어를 실행하고 싶을 때, 파라미터로 함수를 쓰지 않으면 코드가 지나치게 복잡해지기 때문에 함수를 파라미터로 사용할 수 있음
  - 파리미터로서 전달되는 함수를 콜백 함수라고 칭함

- 람다식 : 한줄로 정의가능한 함수를 간단하게 축약한 것
  - 람다식은 디버깅이 불가능

```python
# def minus(x, y):
#     return x - y

minus = lambda x, y : x - y # 람다식은 하나의 수식으로만 리턴값을 정의할 수 있을 때 사용 가능
```

```python
something(10, 20, minus)
```

```python
something(30, 20, lambda x, y : x * y) # 람다식 자체도 파라미터 등으로 사용할 수 있음
```

```python
add_session = requests.Session()

add_session.headers.update({
    "Referer": "",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
})
```

```python
dict1 = {}

def add_find(a, b, c):
    add_url2 = add_url.format(X = a, Y = b)
    try:
        r = session.get(add_url2)
        if r.status_code != 200:
            msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
            raise Exception(msg)
        return json.loads(r.text)
    except Exception as e:
        print("접속에 실패했습니다.")
        print(e)
```

```python
total = len(data_list[:500])
progress = tqdm(total = total)

with futures.ThreadPoolExecutor(max_workers=50) as executor:
    for i in data_list[:500]:
        tar_x = i.split(',')[2]
        tar_y = i.split(',')[3]
        tar_num = i.split(',')[0]

        fu = executor.submit(add_find, tar_x, tar_y, tar_num)

        result = fu.result()
        dict1[int(tar_num)] = result

        fu.add_done_callback(lambda x : progress.update())
```
