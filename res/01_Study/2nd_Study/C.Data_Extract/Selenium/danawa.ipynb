```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
import datetime as dt
```

```python
url_nvidia = 'https://prod.danawa.com/list/?cate=1131480'
url_amd = 'https://prod.danawa.com/list/?cate=1131521'
```

```python
def rip_contents(li):
    prod = []
    for i in li:
        prod_info = {}
        try:
            txt_temp = i.find_elements(by = By.CSS_SELECTOR, value = 'a[name=productName]')[0].text
            prod_info['상품명'] = txt_temp
        except:
            continue

        prod_info['제조사'] = txt_temp[:txt_temp.find(" ")].strip()
        
        try:
            txt_temp = i.find_elements(by = By.CSS_SELECTOR, value = 'div.spec_list')[0].text
            prod_info['칩셋명'] = txt_temp[:txt_temp.find('/')].strip()
        except:
            prod_info['칩셋명'] = None

        try:
            prod_info['가격'] = int((i.find_elements(by = By.CSS_SELECTOR, value = 'p.price_sect strong')[0].text).replace(",", ""))
        except:
            prod_info['가격'] = None

        try:
            prod_info['판매몰'] = int(i.find_elements(by = By.CSS_SELECTOR, value = 'p.chk_sect')[0].text[:-1])
        except:
            prod_info['판매몰'] = None
        
        try:
            txt_temp = i.find_elements(by = By.CSS_SELECTOR, value = 'dl.mt_date dd')[0].text
            txt_temp = dt.datetime.strftime(dt.datetime.strptime(txt_temp, '%Y.%m.'), '%Y-%m-%d')
            prod_info['출시일'] = txt_temp
        except:
            prod_info['출시일'] = None

        if prod_info:
            prod.append(prod_info)

    return prod
```

```python
def danawa_rip(url):
    driver = webdriver.Chrome()
    driver.get(url)

    danawa = []
    idx = 1
    val = '//*[@id="productListArea"]/div/div/div/a[{0}]'

    while True:
        prod_temp = driver.find_elements(by = By.CSS_SELECTOR, value = 'div.main_prodlist_list > ul.product_list > li > div.prod_main_info')
        danawa += rip_contents(prod_temp)
        idx += 1

        try:
            go_next = driver.find_element(by = By.XPATH, value = val.format(idx))
            go_next.click()
            time.sleep(1)
        except:
            try:
                next_page = driver.find_element(by = By.CSS_SELECTOR, value = 'a.edge_nav.nav_next')
                next_page.click()
                idx = 1
                time.sleep(1)
            except:
                break
    
    return danawa
```

```python
danawa_nvidia = danawa_rip(url_nvidia)
```

```python
danawa_amd = danawa_rip(url_amd)
```

```python
geforce = pd.DataFrame(danawa_nvidia)
radeon = pd.DataFrame(danawa_amd)
```
