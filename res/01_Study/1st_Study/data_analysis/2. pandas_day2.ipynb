## 데이터 연결

```python
df1 = pd.read_csv("./data/concat_1.csv")
df2 = pd.read_csv("./data/concat_2.csv")
df3 = pd.read_csv("./data/concat_3.csv")
```

```python
df1.head(1)
```

```python
df2.head(1)
```

```python
concat_df = pd.concat([df1, df2, df3])
```

```python
new_row = pd.Series(["n1", "n2", "n3", "n4"])
```

```python
pd.concat([df1, new_row])
```

- 시리즈에는 열이름이 없기 떄문에 새로운 열로 간주

```python
new_row_df = pd.DataFrame([["n1", "n2", "n3", "n4"]], columns = ["A", "B", "C", "D"])
new_row_df
```

```python
pd.concat([df1, new_row_df])
```

```python
data_dict = {"A" : "n1", "B" : "n2", "C" : "n3", "D" : "n4"}
```

```python
df1.append(data_dict, ignore_index = True)
```

- 딕셔너리는 인덱스가 없기 떄문에 ignore_index가 없으면 에러
- ignore_index = True로 하면 데이터를 연결한 다음 인덱스를 다시 지정

```python
concat_df1 = pd.concat([df1, df2, df3], ignore_index = True)
```

```python
concat_df1
```

## 결측치 대체

```python
ebola.iloc[:10, :5]
```

```python
ebola.fillna(0).iloc[:10, :5]
```

```python
# ffill : 누락값이 나타나기 전의 값으로 채워넣기
ebola.fillna(method = "ffill").iloc[:10, :5]
```

```python
# bfill : 누락값 이후 첫 번째 값으로 앞쪽의 누락값이 변경
ebola.fillna(method = "bfill").iloc[:10, :5]
```

```python
# interpolat() : 누락값 양쪽값의 중간값으로 처리, 데이터가 일정한 간격을 유지하고 있을 때 유용함

ebola.interpolate().iloc[:10, :5]
```
