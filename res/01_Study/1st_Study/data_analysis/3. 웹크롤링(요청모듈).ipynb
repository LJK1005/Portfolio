```python
import requests as rq
```

# requests로 웹페이지 접속

```python
url = "https://www.naver.com"

rq.get(url)
```

```python
rq.post(url)
```

```python
url = "https://www.naver.com/a"

res = rq.get(url)
```

```python
# 응답코드 가져오기
print(res.status_code)
```

```python
# 헤더 가져오기
res.headers
```

```python
url = "https://www.naver.com"

res = rq.get(url)

print(res)
# html 코드 보기
print(res.text)
```

```python
# 쿼리스트링 데이터 생성
res = rq.get(url, params = {"key1" : "value1",
                           "key2" : "value2"})

print(res.url)
```

# urllib으로 웹페이지 접속

```python
from urllib.request import urlopen
```

```python
page = urlopen(url)

print(page)
```

```python
print(page.code) # 응답코드
print(page.headers) # 헤더
print(page.url) # 요청 url
```

```python
print(page.read())
```
