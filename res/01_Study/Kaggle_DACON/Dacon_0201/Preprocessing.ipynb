```python
import pandas as pd
```

```python
df = pd.read_csv("./data/train.csv")
```

```python
df.head()
```

```python
Q = ['질문_1', '질문_2']
A = ['답변_'+str(i) for i in range(1, 6)]
```

```python
df2 = pd.DataFrame()

for i in Q:
    for j in A:
        tmp = df[[i, j]]
        tmp.columns = ['질문', '답변']
        df2 = pd.concat([df2, tmp], axis = 0)
```

```python
df2.to_csv("./data/train_1on1.csv", index = False)
```
