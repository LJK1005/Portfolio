- 훈련횟수별 성적(scatter)

```python
z1 = np.polyfit(df2['tr1'], df3['hrSpd'], 1)
z2 = np.polyfit(df2['tr2'], df3['hrSpd'], 1)
z3 = np.polyfit(df2['tr1']+df2['tr2'], df3['hrSpd'], 1)
```

```python
f1 = np.poly1d(z1)
f2 = np.poly1d(z2)
f3 = np.poly1d(z3)

x1 = df2['tr1']
x2 = df2['tr2']
x3 = df2['tr1']+df2['tr2']

y1 = f1(x1)
y2 = f2(x2)
y3 = f3(x3)
```

```python
fig, ax = plt.subplots(1, 3, figsize = (18, 6))
ax[0].scatter(df2['tr1'], df2['hrSpd'], color = '#62b2de')
ax[1].scatter(df2['tr2'], df2['hrSpd'], color = '#62b2de')
ax[2].scatter(df2['tr2'] + df2['tr1'], df2['hrSpd'], color = '#62b2de')

ax[0].plot(x1, y1, color = '#d46d28')
ax[1].plot(x2, y2, color = '#d46d28')
ax[2].plot(x3, y3, color = '#d46d28')

ax[0].set_title("구보훈련")
ax[1].set_title("습보훈련")
ax[2].set_title("구보+습보")

ax[0].set_xlabel("훈련횟수")
ax[1].set_xlabel("훈련횟수")
ax[2].set_xlabel("훈련횟수")

ax[0].set_ylabel("속력")
ax[1].set_ylabel("속력")
ax[2].set_ylabel("속력")
plt.show()
```
