# 데이터 시각화

- 데이터를 분석한 결과를 쉽게 이해할 수 있도록 표현하여 전달하는 것
- 데이터 시각화가 필요한 이유
    - 많은 양의 데이터를 한 눈에 살펴볼 수 있음
    - 전문 지식이 없더라도 누구나 쉽게 데이터를 인지하고 사용할 수 있음
    - 단순한 데이터의 요약이나 나열보다 더 정확한 데이터 분석 결과를 얻을 수 있음
    - 단순한 데이터에서는 알 수 없었던 중요한 정보를 파악할 수 있음
    
- 파이썬의 시각화 라이브러리
    - matplotlib, seaborn, wordcloud

# 그래프란

- 서로 연관성이 있는 1개 또는 그 이상의 양에 대한 상대 값을 도형으로 나타내는 것

## 데이터의 갯수, 종류에 따른 그래프의 종류

- 일변량(데이터 개수가 1개)
    - 연속형
        - 히스토그램(histogram)
        - 상자 수염 그래프(boxplot)
        - 바이올린 그래프(violin)
        - 커널 밀도 그래프(kernel density curve)
    - 범주형
        - 막대 그래프(bar chart)
        - 파이 그래프(pie chart)
    
- 다변량(데이터 개수가 2개 이상)
    - 연속형
        - 산점도(scater plot)
        - 선 그래프(line)
        - 시계열 그래프
    - 범주형
        - 모자이크 그래프(mosaic graph)
        - Tree Map 그래프

# 그래프 영역 채우기

```python
x = [1, 2, 3, 4]
y = [1, 4, 9 , 16]

plt.plot(x, y)
plt.xlabel("X-Labbel")
plt.ylabel("Y-Label")
plt.fill_between(x[1:3], y[1:3], alpha = 0.5)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9 , 16]

plt.plot(x, y)
plt.xlabel("X-Labbel")
plt.ylabel("Y-Label")
plt.fill_betweenx(y[2:4], x[2:4], alpha = 0.5, color = "pink")
plt.show()
```

# 여러 곡선 그리기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, "r--", a, a**2, "bo", a, a**3, "g-.")
plt.show()
```

```python
plt.plot(a, a, "bo")
plt.plot(a, a**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(a, a**3, color = "springgreen", marker = "^", markersize = 9)
plt.show()
```

# 격자 설정

```python
plt.plot(a, a, "bo")
plt.plot(a, a**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(a, a**3, color = "springgreen", marker = "^", markersize = 9)
plt.grid(True)
plt.show()
```

```python
plt.plot(a, a, "bo")
plt.plot(a, a**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(a, a**3, color = "springgreen", marker = "^", markersize = 9)
plt.grid(True, axis = "y")
plt.show()
```

```python
plt.plot(a, a, "bo")
plt.plot(a, a**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(a, a**3, color = "springgreen", marker = "^", markersize = 9)
plt.grid(True, axis = "y", color = "red", alpha = 0.5, linestyle = "--")
plt.show()
```

# 눈금 표시

```python
plt.plot(a, a, "bo")
plt.plot(a, a**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(a, a**3, color = "springgreen", marker = "^", markersize = 9)

plt.xticks([0, 1, 2])
plt.yticks(np.arange(1,6))

plt.show()
```

```python
plt.plot(a, a, "bo")
plt.plot(a, a**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(a, a**3, color = "springgreen", marker = "^", markersize = 9)

plt.xticks(np.arange(0, 2, 0.2), labels = ["Jan", "", "Feb", "", "Mar", "", "May", "", "June", ""])
plt.yticks(np.arange(0, 7), ("0", "1GB", "2GB", "3GB", "4GB", "5GB", "6GB"))
plt.tick_params(axis = "x", direction = "in", length = 3, pad = 6, labelsize = 14, labelcolor = "green", top = True)
plt.tick_params(axis = "y", direction = "inout", length = 10, pad = 15, labelsize = 12, width = 2, color = "r", rotation = 90)

# 제목 설정
plt.title("Sample graph", loc = "right")
plt.show()
```

- tick_params(): 눈금의 스타일을 설정
- axis : 설정이 적용될 축을 지정. x, y, both 중 선택
- direction : 눈금이 안에서 표시될지 밖에서 표시될지. in, out, inout
- length : 눈금의 길이
- pad : 눈금과 레이블과의 거리
- labelsize : 레이블의 크기
- labelcolor : 레이블의 색상
- top/bottom/left/right : 눈금이 표시될 위치 선택
- width : 눈금의 너비
- color : 눈금의 색상

# 막대 그래프 그리기

```python
x = np.arange(3)
years = ["2017", "2018", " 2019"]
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
plt.show()
```

```python
x = np.arange(3)
years = ["2017", "2018", " 2019"]
values = [100, 400, 900]

plt.bar(x, values, width = 0.6, align = "edge", color = "springgreen",
        edgecolor = "gray", linewidth = 3, tick_label = years, log = True)
plt.xticks(x, years)
plt.show()
```

- width : 막대의 너비
- align : 틱과 막대의 위치, edge로 설정하면 왼쪽 끝에 표시
- color : 막대의 색
- edgecolor : 막대의 테두리 색
- linewidth : 테두리의 두께
- log : y축이 로그 스케일로 표시

# 수평 막대 그래프

```python
y = np.arange(3)
years = ["2017", "2018", " 2019"]
values = [100, 400, 900]

plt.barh(y, values, height = -0.6, align = "edge", color = "springgreen",
        edgecolor = "gray", linewidth = 3, tick_label = years, log = True)
plt.show()
```

# 서브플롯

- plt.subplot(nrow, ncol, pos)
    - 줄, 칸, 위치

```python
x = np.arange(1, 10)
y1 = x * 5
y2 = x * 1

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.show()
```

```python
x = np.arange(1, 10)
y1 = x * 5
y2 = x * 1
y3 = x * 0.3
y4 = x * 0.2

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.hist(np.random.randn(100), bins = 20)
ax2. scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize = [10, 8])

axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.show()
```

# seaborn

```python
tips = sns.load_dataset("tips")
tips
```

## 단변량 히스토그램

```python
ax = plt.subplots()
ax = sns.distplot(tips["total_bill"])
ax.set_title("Total Bill Histogram with Density Plot")
plt.show()
```

```python
ax = plt.subplots()
ax = sns.distplot(tips["total_bill"], kde = True, hist = False)
ax.set_title("Total Bill Histogram with Density Plot")
plt.show()
```

## count 그래프

```python
ax = plt.subplots()
ax = sns.countplot(x = "day", data = tips)
ax.set_title("Count of days")
ax.set_xlabel("Day of the Week")
ax.set_ylabel("Frequency")
plt.show()
```

## 산점도 그래프

```python
ax = plt.subplots()
ax = sns.regplot(x = "total_bill", y = "tip", data = tips)
ax.set_title("scatterplot")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
plt.show()
```
