- 왜 개 사료가 예상보다 빨리 상하게 되는지를 파악
- 이 회사에서 사용하는 5가지 방부제의 사용 용량이 중구난방
- 그 중에서 가장 많은 영향을 끼치는 것이 무엇일까요?<br>

이 회사에서는 먼저 4가지 방부제(A, B, C, D)가 함유된 방부제 묶음을 섞은 뒤 '필러'라는 화학물질로 마무리. 식품과학자들은 A, B, C, D 방부제 중 하나가 문제를 일으키고 있다고 생각, 그게 어떤 것인지 명확하게 "랜덤포레스트"로 어떤 방부제가 사료를 빨리 상하게 하는지 찾아!

- Pres_A : 혼합물의 A비율
- Pres_B : 혼합물의 B비율
- Pres_C : 혼합물의 C비율
- Pres_D : 혼합물의 D비율
- Spoiled : 개 사료가 상했는지 여부를 나타내는 레이블.

**시나리오에서 요구하는 바를 고려하여, 머신러닝을 이용해서 문제를 풀어보세요.**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/dog_food.csv")
```

- 다른 방부제들은 혼합비율에 상관없이 부패된 사료들이 나오는데, C 방부제만 유일하게 9이상 함유됐을때 부패한 사료들이 식별된다. C방부제가 원인인걸로 일단 가설을 세워보고 모델 구축후 결과와 비교해보자.

```python
# 변수지정과 데이터 분리
X = df.drop("Spoiled", axis = 1)
Y = df["Spoiled"]

x_train, x_test, y_train, y_test = train_test_split(X,Y, random_state=4, test_size= 0.25)
```

```python
# 데이터가 잘 나뉘었는지 확인
x_train.shape, x_test.shape
```

```python
#의사결정 나무 모델 돌리기
dt = DecisionTreeClassifier(random_state=4)

dt.fit(x_train, y_train)
```

```python
# 점수 확인
print(dt.score(x_train, y_train))
print(dt.score(x_test, y_test))
```

- 모델 점수가 좋게 나와서 가지치기를 하지 않고, 이대로 시각화를 진행

```python
# 시각화
plt.figure(figsize = (20,15))
plot_tree(dt,max_depth = 7, filled = True, feature_names = ["A", "B", "C", "D"])
plt.show()
```

# 시각화 해석
- 루트 노드는 C함유율이 10.5 이하인지 질문

1. 각 데이터 샘플의 C가 10.5 이하면 왼쪽 가지, 초과면 오른쪽 가지
    - 루트 노드의 총 샘플 수는 367개
    - 이 중에서 음성 클래스(정상 사료)는 265개
    - 양성 클래스(상한 사료)은 102개

2. 왼쪽 노드는 C가 9.5 이하인지 질문
    - yes는 왼쪽 가지, no 는 오른쪽 가지로 이동
    - 노드의 총 샘플 수는 268개
    - 음성 클래스와 양성 클래스 개수는 각각 263개, 5개

3. 오른쪽 노드는 B가 9.5 이하인지 질문
    - 노드의 총 샘플 수는 99개
    - 음성 클래스와 양성 클래스 개수는 각각 2개, 97개

## 특성 중요도

```python
print(dt.feature_importances_)
```

- %순으로 나열하면: C > D > B > A
- C방부제가 부패의 원인으로 보인다.(92.3%)
- C비율이 10.5초과로 늘어날수록 부패한 데이터가 식별된다.

# Bagging

```python
from sklearn.ensemble import BaggingClassifier

bdt = BaggingClassifier(
    base_estimator = DecisionTreeClassifier(), 
    n_estimators = 100,
    max_samples = 0.8,
    oob_score = True,
    random_state = 4)

bdt.fit(x_train, y_train)
print(bdt.score(x_train, y_train))
print(bdt.score(x_test, y_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rdt = RandomForestClassifier(n_estimators = 30)
rdt.fit(x_train, y_train)
print(rdt.score(x_train, y_train))
print(rdt.score(x_test, y_test))
```

# 로지스틱 모델을 이용해서 성능 비교

```python
logi = LogisticRegression()
logi.fit(x_train, y_train)
```

```python
print(logi.score(x_train, y_train))
print(logi.score(x_test, y_test))
```

# 결론

- Decision Tree 모델을 통한 성능은 Bagging, RandomForest 둘다 적용해도 차이가 없었다.
- Logisitic model을 통해서 테스트해봐도 train data score만 낮아질뿐이다.
- 하지만, 성능비교에 score말고 다른 metrics도 봐야지 알것 같다.

- C방부제가 문제다.
