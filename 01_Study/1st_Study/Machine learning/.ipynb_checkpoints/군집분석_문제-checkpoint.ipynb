# 군집분석

한 기업이 해킹을 당해서 우리에게 도움을 요청했습니다! 다행스럽게도 그 기업의 포렌식 기술자들은 session time, location, wpm typing speed 등과 같은 해킹에 대한 귀중한 정보를 수집했습니다. 담당 포렌식 기술자와의 회의에서 듣기로는 해커들이 서버에 접속하는데에 사용한 각 세션의 메타 데이터들을 수집한 것이라고 합니다.

다음은 수집된 데이터입니다

- Session_Connection_Time : 세션이 지속된 시간(분)입니다.
- Bytes Transferred : 세션 중에 전송된 데이터 용량(MB)입니다.
- Kali_Trace_Used : 해커의 Kali Linux 사용 여부입니다.
- Servers_Corrupted : 공격으로 손상된 서버 수 입니다.
- Pages_Corrupted : 무단으로 접근한 페이지 수 입니다.
- Location : 공격이 시작된 위치입니다.(해커들이 VPN을 사용했기 때문에 아마 쓸모가 없을 것 같습니다.)
- WPM_Typing_Speed : 세션 로그를 기준으로 예상한 타이핑 속도입니다.

이 회사에서는 3명의 해커를 의심 중입니다. 2명의 해커는 어느정도 확신하고 있지만 세 번째 해커가 해킹에 연루되었는지 아직 확실하지 않습니다. 우리에게 의뢰한 부분은 이 부분입니다. 즉, 이번 해킹 사건에서 2명이 공격을 한 것인지 3명이 공격을 한 것인지 분석해달라고 합니다. 확실하지는 않지만 클러스터링으로 알아 볼 수 있을 것 같습니다.

마지막으로 한가지 알려드릴 중요한 사실은, 각 해커들의 공격횟수는 거의 동일합니다. 예를 들어, 총 공격 횟수가 100번 이었다면, 50건씩 2명이 공격을 했거나 3명이 약 33건씩 공격을 했다는 말입니다. 데이터는 hack_data.csv 파일로 전달해 드리겠습니다.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("./data/hack_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
# VPN으로 인하여 Location은 의미가 없으므로 삭제
df.drop(["Location"], axis = 1, inplace = True)
```

```python
df.head()
```

# 데이터 시각화

```python
sns.pairplot(df)
```

- 시각화 데이터로 보았을 때 클러스터 수(=해커 수)는 2로 추정됨

# 이너셔 그래프

- 이너셔 그래프를 이용하여 적합한 k값 추정

```python
inertia = []

for k in range(1, 7):
    km = KMeans(n_clusters = k, random_state = 4, n_init = 10)
    km.fit(df)
    inertia.append(km.inertia_)

plt.plot(range(1, 7), inertia)
plt.xlabel("k")
plt.ylabel("inertia")
plt.show()
```

- 클러스터 개수가 3을 초과할때 이너셔의 기울기 감소율이 둔화됨
    - 그래프 상으로는 클러스터 개수가 2일 경우가 최적의 클러스터 개수로 보임

# 클러스터 개수가 2인 모델과 3인 모델 비교

- 전제사항에서 해커별로 공격한 횟수는 거의 비등하기 때문에 클러스터당 요소 개수를 파악하여 적합한 클러스터 수인지 파악

```python
km2 = KMeans(n_clusters=2, random_state=1)
km3 = KMeans(n_clusters=3, random_state=1)
```

```python
km2.fit(df)
km3.fit(df)
```

```python
pd.value_counts(km2.labels_)
```

```python
pd.value_counts(km3.labels_)
```

- 클러스터의 개수가 2일 경우가 3인 경우보다는 클러스터 요소의 개수가 비교적 고르게 나타나고 있음
    - 요소 개수별로 파악했을 경우에는 클러스터 개수(=해커 수)는 2인 것으로 보임

# 클러스터별 분산 확인

- 전제사항에서 클러스터(= 해커)별 공격 횟수가 고르다는 사실에 착안하면 각 클러스터 중심 거리에서 각 요소까지의 거리를 오름차순으로 나열한 후, 클러스터별 요소 개수(= 전체 데이터 / 클러스터 개수)로 자른 뒤 그 분산을 구한다면 그 값들이 어느정도 유사해야 함
    - 해당 사항을 클러스터 개수가 2인 모델과 3인 모델간에 비교
    - k-means 모델에서 클러스터별 요소 개수를 강제로 통일시킬수 있는 패러미터가 없어 아래 방법으로 진행, 정확한 추론 방법은 아님 (중복 선택되는 요소가 있을수 있으며, 각 클러스터(해커) 특성상 분산이 제각각일수도 있음)

```python
# 클러스터별 거리 확인
km2.transform(df)
```

```python
d2 = km2.transform(df)
```

```python
# 단일 클러스터에 대한 전체 요소의 거리
d2[:, 0]
```

```python
# 해당 거리 데이터를 오름차순으로 나열 후 요소 수로 자르기
np.sort(d2[:, 0])[0:167]
```

```python
# 각 클러스터별 분산 확인
print(np.var(np.sort(d2[:, 0])[0:167]))
print(np.var(np.sort(d2[:, 1])[0:167]))
```

```python
# 클러스터 3개 모델의 클러스터별 분산 확인
d3 = km3.transform(df)
print(np.var(np.sort(d3[:, 0])[0:111]))
print(np.var(np.sort(d3[:, 1])[0:111]))
print(np.var(np.sort(d3[:, 2])[0:111]))
```

- 클러스터 개수가 3인 모델의 경우 특정 클러스터(1번)에서만 분산이 크게 다른 모습을 보임
    - 클러스터 개수(= 해커 수)가 3인 추론은 적합하지 않을 수 있음
    
- 위 내용들로 추론한 결과 해커의 수는 2와 3중에서는 2가 더욱 적합한 것으로 보임
