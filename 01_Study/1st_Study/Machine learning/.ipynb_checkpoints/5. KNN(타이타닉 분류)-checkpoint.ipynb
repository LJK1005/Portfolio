```python
import pandas as pd
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

# 데이터 준비

```python
# Load_dataset 함수를 사용하여 데이터프레임으로 변환
df = sns.load_dataset("titanic")
df.head()
```

# 데이터 탐색

```python
df.shape
```

```python
df.info()
```

```python
df.dtypes
```

```python
# 결측치 개수 확인
df.isnull().sum()
```

```python
# 결측치가 많은 deck열은 삭제
# embarked와 내용이 겹치는 embark_town 열을 삭제
rdf = df.drop(["deck", "embark_town"], axis = 1)
rdf.columns.values
```

```python
# age 열예 결측치가 있는 모든 행을 삭제(177개 행)
rdf = rdf.dropna(subset = ["age"], how = "any", axis = 0)
```

```python
len(rdf)
```

```python
# embarked 열의 NaN 값을 승선도시 중에서 가장 많이 출현한 값으로 치환하기 (최빈값 대체)
most_freq = rdf["embarked"].value_counts().idxmax()
```

```python
# 최빈값을 확인하는 다른 방법
rdf.describe(include = "all")
```

```python
rdf["embarked"].fillna(most_freq, inplace = True)
```

# 분석에 사용할 속성을 선택

```python
ndf = rdf[["survived", "pclass", "sex", "age", "sibsp", "parch", "embarked"]]
ndf.head()
```

```python
# 원핫인코딩 - 범주형데이터를 인식할 수 있도록 숫자형으로 변환
onehot_sex = pd.get_dummies(ndf["sex"])
```

```python
onehot_sex
```

```python
ndf = pd.concat([ndf, onehot_sex], axis = 1)
```
