```python
pp = YjkPreprocessor()
pp.set_strategy(outline = None, null = None)
```

```python
x_train, x_test, y_train, y_test = my_train_test_split(df, yname = '분기당_매출_금액', test_size = 0.2, random_state = 0)
```

```python
train = pd.concat([x_train, y_train], axis = 1)
```

```python
sum_df = train.groupby('서비스_업종_코드').sum(numeric_only = True)
```

```python
atv = sum_df['분기당_매출_금액'] / sum_df['분기당_매출_건수']
```

```python
# x_train['ATV'] = x_train['서비스_업종_코드'].apply(lambda x : atv[x])
# x_test['ATV'] = x_test['서비스_업종_코드'].apply(lambda x : atv[x])

df['ATV'] = df['서비스_업종_코드'].apply(lambda x : atv[x])
```

```python
df_p = pp.fit_transform(df, yname = '분기당_매출_금액')
```

```python
x_train, x_test, y_train, y_test = my_train_test_split(df_p, yname = '분기당_매출_금액', test_size = 0.2, random_state = 0)
```

```python
support = YjkRegressorSupport()
```

```python
model = support.call_model('SGDRegressor')
```

```python
model.set_params(early_stopping = False, random_state = 10)
```

```python
train_li = []
test_li = []
```

```python
for _ in range(100):
    model.partial_fit(x_train, y_train)
    train_li.append(model.score(x_train, y_train))
    test_li.append(model.score(x_test, y_test))
```

```python
plt.figure(figsize = (10, 5))
plt.plot(train_li, label = 'train')
plt.plot(test_li, label = 'test')
plt.xlabel("epochs")
plt.show()
```
