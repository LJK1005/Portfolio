- 확률적 경사하강법 (SGD) : 손실함수의 기울기가 최수가 되게 하는 지점을 찾는 경사하강법
    - 손실함수 : 기울기와 절편을 찾는 과정에서 대입한 값이 실제 값과 최대한 근접하도록 적용하는 함수
        - 회귀 : MAE, MSE, RMSE
        - 분류 : binary crossentropy(이항 교차 엔트로피, 이진분류), (Sparse) Categorical Crossentropy(범주형 엔트로피, 다항분류)
    - 경사하강법 : 방정식의 근사값을 발견하기 위한 최적화 알고리즘 
        - 경사하강법은 미분을 통하여 손실이 최소가 되는 지점을 찾음
    - Optimizer : 손실을 최소화시키기 위해 경사를 내려가는 방법을 찾는 알고리즘
        -  GD : 모든 값을 일일히 대입 (느림)
        -  SGD : GD에서 대입할 샘플을 일부 생략하여 빠르게 계산을 수행 (다만 무작위)
            - 노이즈 : 랜덤으로 데이터를 선택함으로 인하여 발생하는 오차들
