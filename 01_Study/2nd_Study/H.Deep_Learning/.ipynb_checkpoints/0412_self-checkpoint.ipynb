```python
from sklearn.datasets import load_iris
```

```python
X, Y = load_iris(return_X_y = True)
```

```python
Y
```

```python
model = Sequential(layers = [
    Dense(32, activation = 'relu', input_shape = (4,)),
    Dropout(0.15),
    Dense(8, activation = 'relu'),
    Dropout(0.15),
    Dense(3, activation = 'softmax')
])
```

```python
es = EarlyStopping(monitor = 'val_acc', patience = 30)
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy', 'sparse_categorical_crossentropy'])
```

```python
history = model.fit(X, Y, epochs = 300, verbose = 0, validation_split = 0.2)
```

```python
tmp = pd.DataFrame(history.history)
tmp
```

```python
yjk_history_plot(history, metrics = 'accuracy')
```

```python
df = pd.read_excel("./data/boston.xlsx")
```

```python
df.head()
```

```python
df_p = pp.fit_transform(df, yname = 'MEDV')
```

```python
X = df_p.copy()
Y = X.pop('MEDV')
```

```python
model = Sequential(layers = [
    Dense(64, activation = 'relu', input_shape = (13,)),
    Dropout(0.2),
    Dense(32, activation = 'relu'),
    Dropout(0.15),
    Dense(1, activation = 'linear')
])
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mae', metrics = 'mse')
```

```python
es = EarlyStopping(patience = 10, restore_best_weights = True)
```

```python
history = model.fit(X, Y, validation_split = 0.2, epochs = 1000, callbacks = [es])
```

```python
yjk_history_plot(history, metrics = 'mse')
```
