````python
url2 = 'https://browse.gmarket.co.kr/search?keyword=%ED%94%84%EB%A6%B0%ED%84%B0&k=30&p={0}'
````

````python
session2 = requests.Session()

session2.headers.update({
    "Referer": "",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
})
````

````python
def gmarket_rip(url):
    try:
        r2 = session2.get(url)
            
        if r2.status_code != 200:
            msg2 = "[%d Error] %s 에러가 발생함" % (r2.status_code, r2.reason)
            raise Exception(msg2)
    except Exception as e:
        print("접속에 실패했습니다.")
        print(e)

    r2.encoding = "utf-8"

    soup3 = BeautifulSoup(r2.text)
    contents = soup3.select("div.box__item-container")

    gmarket = []

    for i in contents:
        try:
            item_title = i.select('span.text__item')[0].text
        except:
            item_title = None

        try:
            seller = i.select('span.text__seller')[0].text
        except:
            seller = None

        try:    
            origin_price = int(i.select('div.box__price-original span.text__value')[0].text.replace(",", ""))
        except:
            origin_price = None

        try:
            discount_rate = i.select('div.box__discount span.text__value')[0].text
        except:
            discount_rate = None

        try:
            discount_price = int(i.select('strong.text__value')[0].text.replace(",", ""))
        except:
            discount_price = None

        gmarket.append({'상품명':item_title, '판매자':seller, '원가격':origin_price, '할인율':discount_rate, '판매가':discount_price})

    return gmarket
````

````python
gmarket_all = []

for i in range(1, 11):
    url_g = url2.format(i)
    gmarket_all += gmarket_rip(url_g)
````

````python
with open('gmarket_extract.csv', 'w', encoding = 'utf-8')as f:
    for i, v in enumerate(gmarket_all):
        if i == 0:
            f.write(','.join(list(v.keys()))+'\n')
        f.write(f'"{v["상품명"]}","{v["판매자"]}","{v["원가격"]}","{v["할인율"]}","{v["판매가"]}"\n')
````
