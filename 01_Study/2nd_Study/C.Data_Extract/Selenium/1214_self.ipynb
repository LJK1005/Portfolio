- 셀레니움 관련 패키지
  - selenium : 웹 브라우저를 제어하는 파이썬 패키지
  - chromedriver : 셀레니움과 Chrome을 연결하는 프로그램
    - chromedriver_autoinstaller : 현재 컴퓨터에 설치된 크롬 버전에 맞는 크롬드라이버를 자동으로 다운받는 패키지

```python
import chromedriver_autoinstaller
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.by import By
```

```python
chromedriver_autoinstaller.install()
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(5)
```

- 수집할 페이지로 이동

```python
driver.get("https://data.hossam.kr/py/image.html")
```

- 셀레니움을 통한 크롬 브라우저 제어

1. 키보드 입력 전달
   - 특정 HTML요소 가져오기
     - 검색어를 입력할 수 있는 요소

```python
query = WebDriverWait(driver, 3).until(lambda x:x.find_element(By.CSS_SELECTOR, "#query"))
query
```

- 문자열을 해당 요소에 입력하기

```python
# 기존에 있던 입력값을 제거
query.clear()

# 입력 진행, 입력값이 누적되는 방식
query.send_keys("파이썬")
```

2. 마우스 클릭 전달
    - 검색 버튼 가져오기

```python
button = WebDriverWait(driver, 3).until(lambda x:x.find_element(By.CSS_SELECTOR, "#searchForm button"))
button
```

- 검색 버튼 클릭

```python
button.click()
```

3. 스크롤 이동
  - 크롬브라우저로 하여금 Javascript 코드를 실행하도록 처리
  - "window.scrollTo(0, document.body.scrollHeight);" : 화면이 맨 아래로 이동하는 자바스크립트 구문

```python
for i in range(0, 10): # n회 반복
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # 화면을 맨 아래로 이동
    time.sleep(0.5) # 내려간 스크롤로 인하여 컨텐츠가 새롭게 로딩될때 까지 일정시간 대기
```

4. 브라우저상에 나타나는 소스코드 가져오기
   - 해당 코드는 크롬브라우저의 개발자도구로 확인되는 내용과 동일함

```python
driver.page_source
```

- bs4에 해당 내용 전달

```python
soup = BeautifulSoup(driver.page_source)
```

- 브라우저 닫기

```python
try:
    driver.close()
except:
    pass
```

- bs4 활용으로 HTML코드 분석

```python
img = soup.select("li > a > img")
```

```python
for i in img:
    print(i.attrs['src'])
```
