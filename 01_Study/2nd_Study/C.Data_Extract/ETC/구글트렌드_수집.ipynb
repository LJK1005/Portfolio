## #01. 준비과정

### [1] 패키지 참조

```python
from pytrends.request import TrendReq
```

### [2] pytrends 객체 생성

| 파라미터 | 설명 |
|---|---|
| `hl` | host language |
| `tz` | timezone (분단위 설정) |

우리나라는 `GMT+9`이므로 9시간에 60을 곱한 값을 설정

```python
pytrends = TrendReq(hl='ko', tz=540)
```

## #02. 트렌드 검색

### 검색어 설정

```python
keywords = ['sklearn', 'tensorflow', 'pytorch']
```

### 검색 설정

| 파라미터 | 설명 |
|---|---|
| `cat` | 카테고리. `0`인 경우 전체. 자세한 항목은 `pytrends.categories()`의 결과를 참고 |
| `timeframe` | 검색 시작일, 검색 종료일 |
| `geo` | 검색 지역 |
| `gprop` | 이미지, 뉴스, 유튜브, froogle(구글쇼핑)으로 세분화 할 경우 설정. 기본값은 그냥 구글 검색 |

```python
pytrends.build_payload(keywords, cat=0, 
			timeframe='2023-01-01 2023-12-15', 
            geo='KR', gprop='')
```

### 본격적인 데이터 탐색

```python
# 설정한 기간 동안의 트렌드 지수
data = pytrends.interest_over_time()
data
```

### 지역별 트렌드 지수

```python
pytrends.interest_by_region(
  		resolution='COUNTRY', # 시, 나라, 지하철..?, 지역 레벨로 설정 가능
  		inc_low_vol=True, # 검색량이 많지 않은 지역도 포함할건지
  		inc_geo_code=True # 국가명하고 iso코드 같이 표시할건지
) 
```
