- 한글 사용하기

```python
import os
from matplotlib import font_manager
```

```python
font_list = font_manager.findSystemFonts() # 시스템에 설치된 폰트 가져오기
font_list.sort() # 정렬

fonts = []

for i in font_list:
    fp = font_manager.FontProperties(fname = i) # 각 폰트의 정보를 불러옴

    try:
        font_name = fp.get_name() # 폰트의 실제 이름 가져오기
        # print(font_name) # 폰트 이름 출력
        fonts.append(font_name)
    except:
        continue
```

```python
# 폰트 설정
plt.rcParams["font.family"] = 'Malgun Gothic'

# 글자 크기
plt.rcParams["font.size"] = 12

# 그래프 크기 설정(inch)
plt.rcParams["figure.figsize"] = (8, 5)

# 유니코드에서 음수 부호 설정
plt.rcParams["axes.unicode_minus"] = False
```

```python
plt.figure()
plt.title("한글 제목")
plt.plot([-20, -10, 0, 10, 20], [1, 2, 3, 4, 5])
plt.xlabel("엑스축 라벨")
plt.ylabel("와이축 라벨")
plt.grid()
plt.show()
plt.close()
```

- 영역별 개별 폰트 설정
  - fontdict 파라미터를 이용

```python
font1 = {'family' : 'Batang', 'size' : 12, 'color' : '#0066ff', 'weight' : 'bold'}
font2 = {'family' : 'Gulim', 'size' : 14, 'color' : '#3322aa'}
font3 = {'family' : 'Gulim', 'size' : 12, 'color' : '#bb5033', 'weight' : 'bold'}

plt.figure()
plt.title("한글 제목", fontdict = font1)
plt.plot([-20, -10, 0, 10, 20], [1, 2, 3, 4, 5])
plt.xlabel("엑스축 라벨", fontdict = font2)
plt.ylabel("와이축 라벨", fontdict = font3)
plt.grid()
plt.show()
plt.close()
```

- 복수 그래프 그리기

```python
plt.figure()
plt.title("복수의 plot")

plt.plot([1, 3, 5, 7, 9], "r--")
plt.plot([10, 15, 20, 25, 30], "g:s")

plt.grid()
plt.show()
plt.close()
```

- 범례(legend)

```python
plt.figure()
plt.title("복수의 plot")

plt.plot([1, 3, 5, 7, 9], "r--", label = "1번")
plt.plot([10, 15, 20, 25, 30], "g:s", label = "2번")

plt.grid()
plt.legend(loc = 'lower right')
plt.show()
plt.close()
```
